```python
from tkinter import *
from tkinter import filedialog as fd

win = Tk()
win.geometry('200x200')

text=Text(win, height=12)
text.grid(row=0,column=0)

def select_file():
    types=(('text file', '*.txt'),('ALL files', '*.*'))
    f = fd.askopenfile(title="open a file", initialdir='C:/', filetypes=types)
    print(f)

    text.insert('1.0',f.read())

btn1 = Button(win, text="open file", command= select_file)
btn1.grid(row=1,column=0)

win.mainloop()
```
